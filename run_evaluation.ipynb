### Chạy hypersearch 
Dùng để tự động thử nhiều cấu hình đánh giá khác nhau (expand, fill-mode, mode, coef,...) và tìm ra thông số nào cho kết quả tốt nhất.\
Tối ưu hóa F1-score bằng cách thử nhiều kết hợp:
- expand: 0.05, 0.1, 0.15, 0.2,...

- fill_mode: linear, nearest, nearest-up

- mode: frame hoặc shot

- coef: số lượng keyframe lấy thêm

```python
import sys

# === THIẾT LẬP CÁC BIẾN ĐƯỜNG DẪN ===
groundtruth_folder = "/mnt/d/Computer Vision/Project/TAC-SUM/evaluation/data"
summary_folder = "/mnt/d/Computer Vision/Project/TAC-SUM/Test/summaries"

# === CÁC THAM SỐ TÌM KIẾM ===
mode = "shot"
min_coef = "0.0"
max_coef = "0.1"
iter_coef = "0.5"
min_expand = "0.15"
max_expand = "0.30"
iter_expand = "0.05"

# Thêm thư mục chứa hypersearch.py
sys.path.append("/mnt/d/Computer Vision/Project/TAC-SUM/evaluation")

from hypersearch import search

# Gán sys.argv để giả lập dòng lệnh
sys.argv = [
    "hypersearch.py",

    "--groundtruth-folder", groundtruth_folder,
    "--summary-folder", summary_folder,

    "--mode", mode,

    "--min-coef", min_coef,
    "--max-coef", max_coef,
    "--iter-coef", iter_coef,

    "--min-expand", min_expand,
    "--max-expand", max_expand,
    "--iter-expand", iter_expand,
]

# Chạy tìm kiếm
search() 
```
